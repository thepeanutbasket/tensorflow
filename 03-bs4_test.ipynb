## 수집한 웹 컨텐츠에서 텍스트 추출

```python
import Http
from bs4 import BeautifulSoup
```

```python
text = Http.get_url("https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=103&oid=262&aid=0000014128", encoding = 'euc-kr')
text
```

![nn](KakaoTalk_20210116_110213935.jpg)

```python
# 수집한 웹 컨텐츠를 bs4 객체로 변환
soup = BeautifulSoup(text, 'html.parser')
```

```python
# 기사 제목 추출 -> id 속성을 같는 경우 '#'를 붙여 전달
# select() 함수의 리턴형은 무조건 list로 들어간다.
title_list = soup.select('#articleTitle')
title_list
```

```python
# 추출된 영역에서 text만 뽑기
title = title_list[0].text
title
```

## 본문 영역 얻기

```python
body_list = soup.select('#articleBodyContents')
body_list
```

## 본문에서 불필요한 태그 영역을 제거

```python
body_tag = body_list[0]

for target in body_tag.find_all('script'):
    target.extract()
    
for target in body_tag.find_all('b'):
    target.extract()
    
for target in body_tag.find_all('table'):
    target.extract()
    
for target in body_tag.find_all('div'):
    target.extract()

for target in body_tag.find_all('br'):
    target.extract()


body_tag
```

```python
body = body_tag.text.strip()
body
```
