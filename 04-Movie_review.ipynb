## 모든 리뷰 페이지의 리뷰 가져오기

```python
reviews = []

url = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=163834&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page=%d'

# 페이지 번호
page = 1

# 반복 횟수를 가늠할 수 없으므로 무한루프
while True:
    review_url = url % page
    
    text = Http.get_url(review_url)
    
    # 수집한 웹 컨텐츠를 bs4 객체로 변환
    soup = BeautifulSoup(text, 'html.parser')
    
    # 리뷰 리스트 추출 --> class로 접근
    # 계층 구조는 ' > '로 표기 ==> .score_reple 클래스 안에 p 가 있고 그 안에 span이 있어서 그걸 갖고와라
    review_list = soup.select('.score_reple > p > span[id]')
    
    for item in review_list:
        reviews.append(item.text.strip())
        
    # 반복 1회 수행시마다 page 번호를 증가
    page += 1
    
    # 전체 리뷰 수
    total = soup.select('.total > em')[0].text.strip()
    total = int(total.replace(',',''))
    
    print('%d/%d 수집됨' % (len(reviews), total))
          
    if len(reviews) >= total:
          break
```

```python
print('수집된 전체 리뷰 수는 %d건입니다.' %len(reviews))
```

```python
with open('result.txt', 'w', encoding = 'utf-8') as f:
    for line in reviews:
        f.write(line.strip() + '\n')
```
